#### 2. Basic Graph

Documentation: *https://matplotlib.org/stable/api/pyplot_summary.html*

**Line Chart**

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize our graph (should be done at the very beginning!)
#figsize is kind of its ratio
#dpi: pixels per inch
plt.figure(figsize=(5,3), dpi=500) 

#give our graph a title
plt.title("Our First graph!",
          fontdict={"fontname": "Comic Sans MS","fontsize":20})

#give the xlabel
plt.xlabel("X")

#give the ylabel
plt.ylabel("Y")

#determine x axis ticks
plt.xticks([0,1,2,3])

#determine y axis ticks
plt.yticks([0,2,4,6,8,10])

# plot need 2 lists for x and y
# label is the data name
# color is the line color, can be hexadecimal also

#keyword argument notation
# plt.plot(x,y, label="2x", color="red",linewidth=2,marker=".",markersize=10,markeredgecolor='blue')

# shortcut notation
# [color][marker][line]
plt.plot(x,y,"r^-", label="2x")

# Plot another 2 lines!
x2 = np.arange(0,4,0.5)
plt.plot(x2[:5], x2[:5]**2, "b.-", label="x**2")
plt.plot(x2[4:], x2[4:]**2, "g.--")

#add legend (it depends on label we pass into plt.plot)
plt.legend()

#save graph into png
plt.savefig("mygraph.png", dpi=100)

#display the graph!
plt.show()
```

**Bar Chart**

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

#resize
plt.figure(figsize=(6,4))

#draw a bar chart!
bars = plt.bar(labels, values)

#draw hatches in each bar
bars[0].set_hatch('/')
bars[1].set_hatch('.')
bars[2].set_hatch('*')

#give a title
plt.title("First Bar Chart")

#show the chart
plt.show()
```

```python
x = ["C++", "C#", "Python", "Java", "Go"]
y = [20, 50, 140, 1, 45]

plt.bar(x,y, color="green", width=0.7)

plt.title("Programming Language Popularity")

plt.show()
```

**Scatter Chart**

```python
X_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

print(X_data)

plt.scatter(X_data, y_data, 
            color='blue', 
            marker='*',
            s=100,
            alpha=0.8)

plt.show()
```

**Histogram**

```python
#create np of random number, generated by normal distribution
ages_norm = np.random.normal(loc=20, #mean
                        scale=1.5, #std deviation
                        size=1000 #data size
                        )

ages_uni = np.random.uniform(low=10,
                             high=50,
                             size=1000
                             )

#bins params = how many blocks you want to plot?
#cumulative will plot cumulative chart
plt.hist(ages_norm, color="cyan", bins=20, cumulative=True)
plt.show()

plt.hist(ages_uni, color="magenta", bins=20)
plt.show()
```

**Plot Multiple Figures**

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

#plot 2 independent figures
plt.figure(1)
plt.scatter(x1, y1)

plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

**Subplots**

```python
x = np.arange(100)

#fig type is Figure object
#axis type is Axes object (can be list of Axes)
fig, axis = plt.subplots(nrows=2, 
                         ncols=2,
                         figsize=(15,10)
                         )

#no subplots will overlap others, automatically
fig.tight_layout()

#adjust the entire figure, manually
fig.subplots_adjust(left=0.125,
                    right=0.9,
                    bottom=0.1,
                    top=0.9,
                    wspace=0.3, #horizontal gap
                    hspace=0.5 #vertical gap
                    )

#give super title for our main figure
fig.suptitle("My First Subplots!!!")

#draw graph on upper left subplots
axis[0,0].plot(x, np.sin(x))
axis[0,0].set_title("Sine Wave")

#draw graph on upper right subplots
axis[0,1].plot(x, np.cos(x))
axis[0,1].set_title("Cosine Wave")

axis[1,0].plot(x, np.random.random(100))
axis[1,0].set_title("Pure Random Function")

axis[1,1].plot(x, np.log(x))
axis[1,1].set_title("Log Function")

#export CURRENT FIGURE as an image
plt.savefig("fourplots.png", dpi=100)

plt.show()

#savefig after show doesn't work, i don't know why
```

```python
x = np.linspace(0,4,1000)

#or using add_subplot method also works~
fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))

plt.show()
```

**3.2 FIFA Data**

```python
fifa = pd.read_csv("fifa_data.csv")

#print top 5 list
fifa.head(5)
```

**3.2.1 Histogram**

```python
bins = [x for x in range(0, 101, 10)]

#Plot histogram
#bins is like saying "how wide is our rectangle?"
plt.hist(fifa.Overall, bins=bins, color="#abcdef")

plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Skill level")

plt.title("Distribution of Player Skills in FIFA 2018")

plt.show()
```

**3.2.2 Pie Charts**

```python
#DataFrame.loc is to filter dataframe based on certain condition, return a dataframe
#DataFrame.count() count non-null cells on every row/column, return a Series
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

#plot the pie chart, passing List of numbers
plt.pie([left, right], 
        labels=["Left Foot", "Right Foot"], 
        autopct='%.2f %%',
        colors=["#abcdef", "#aabbff"])

plt.title("Foot Preference of FIFA Players")

plt.show()
```

```python
#Playing with player's weight
#Weight stored in string, so convert it first to integer
weightInt = []
for weight in fifa.Weight:
    if(type(weight) == str):
        weightInt.append(int(weight.strip('lbs')))
    else:
        weightInt.append(weight)

fifa.Weight = weightInt

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

plt.style.use("ggplot")
plt.pie([light, light_medium, medium, medium_heavy, heavy],
        labels=["Under 125lbs", "125-150lbs", "150-175lbs", "175-200lbs", "Over 200lbs"],
        autopct='%.2f %%',
        explode=(.4,.2,0,0,.4) #slice the chart
        )

plt.show()
```

```python
#print column 'Overall' where Club == 'FC Barcelona'
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

#boxplot to analyze 1st quartile, median, and 3rd quartile of data
plt.boxplot([barcelona, madrid, revs],
            labels=['FC Barcelona', 'Real Madrid', 'New England Revolution'])

plt.show()
```

**Scatter Plot**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ax = plt.axes(projection="3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_title("3D Plot")
plt.show()
```

**Line Plot**

```python
ax = plt.axes(projection="3d")

x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
ax.set_title("Just a Random Line")
plt.show()
```

**Surface Plot**

```python
ax = plt.axes(projection="3d")

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

#what meshgrid does?
#for vectors x1, x2, ..., xn as parameters,
#Ni = len(xi)
#returns (N1, N2, N3, ..., Nn) shaped arrays
#since len(x) and len(y) = 100,
#X.shape and Y.shape = (100,100)
X, Y = np.meshgrid(x,y)
print('X shape: {}, Y shape: {}'.format(X.shape, Y.shape))

Z = np.sin(X) * np.cos(Y)

#Plot the surface
ax.plot_surface(X, Y, Z, cmap="Spectral")

```
