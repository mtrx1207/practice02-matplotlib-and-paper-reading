# Learn Matplotlib

#### 1. Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### 2. Basic Graph

Documentation: *https://matplotlib.org/stable/api/pyplot_summary.html*

**Line Chart**

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize our graph (should be done at the very beginning!)
#figsize is kind of its ratio
#dpi: pixels per inch
plt.figure(figsize=(5,3), dpi=500) 

#give our graph a title
plt.title("Our First graph!",
          fontdict={"fontname": "Comic Sans MS","fontsize":20})

#give the xlabel
plt.xlabel("X")

#give the ylabel
plt.ylabel("Y")

#determine x axis ticks
plt.xticks([0,1,2,3])

#determine y axis ticks
plt.yticks([0,2,4,6,8,10])

# plot need 2 lists for x and y
# label is the data name
# color is the line color, can be hexadecimal also

#keyword argument notation
# plt.plot(x,y, label="2x", color="red",linewidth=2,marker=".",markersize=10,markeredgecolor='blue')

# shortcut notation
# [color][marker][line]
plt.plot(x,y,"r^-", label="2x")

# Plot another 2 lines!
x2 = np.arange(0,4,0.5)
plt.plot(x2[:5], x2[:5]**2, "b.-", label="x**2")
plt.plot(x2[4:], x2[4:]**2, "g.--")

#add legend (it depends on label we pass into plt.plot)
plt.legend()

#save graph into png
plt.savefig("mygraph.png", dpi=100)

#display the graph!
plt.show()
```

**Bar Chart**

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

#resize
plt.figure(figsize=(6,4))

#draw a bar chart!
bars = plt.bar(labels, values)

#draw hatches in each bar
bars[0].set_hatch('/')
bars[1].set_hatch('.')
bars[2].set_hatch('*')

#give a title
plt.title("First Bar Chart")

#show the chart
plt.show()
```

**3.1 Line Graph of Gas Price**

```python
gas = pd.read_csv("gas_prices.csv")

#try what we got!
print(gas)
```

```python
#Now print the data based on column name
print(type(gas.USA)) #the type is pandas' Series
print(gas.USA)
```

```python
#Plot the data!
plt.figure(figsize=(12,8))

plt.title("Gas Prices over Time (USD)")

# plt.plot(gas.Year, gas.USA, "b.-", label="USA")
# plt.plot(gas['Year'], gas.Canada, "r.-", label="Canada")
# plt.plot(gas['Year'], gas['South Korea'], "g.-", label="South Korea")

#plot all
for country in gas:
    if(country != 'Year'):
        plt.plot(gas['Year'], gas[country], label=country)

plt.xticks(gas.Year[::2])

plt.xlabel("Year")
plt.ylabel("USD")

plt.legend()

plt.show()
```

**3.2 FIFA Data**

```python
fifa = pd.read_csv("fifa_data.csv")

#print top 5 list
fifa.head(5)
```

**3.2.1 Histogram**

```python
bins = [x for x in range(0, 101, 10)]

#Plot histogram
#bins is like saying "how wide is our rectangle?"
plt.hist(fifa.Overall, bins=bins, color="#abcdef")

plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Skill level")

plt.title("Distribution of Player Skills in FIFA 2018")

plt.show()
```

**3.2.2 Pie Charts**

```python
#DataFrame.loc is to filter dataframe based on certain condition, return a dataframe
#DataFrame.count() count non-null cells on every row/column, return a Series
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

#plot the pie chart, passing List of numbers
plt.pie([left, right], 
        labels=["Left Foot", "Right Foot"], 
        autopct='%.2f %%',
        colors=["#abcdef", "#aabbff"])

plt.title("Foot Preference of FIFA Players")

plt.show()
```

```python
#Playing with player's weight
#Weight stored in string, so convert it first to integer
weightInt = []
for weight in fifa.Weight:
    if(type(weight) == str):
        weightInt.append(int(weight.strip('lbs')))
    else:
        weightInt.append(weight)

fifa.Weight = weightInt

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

plt.style.use("ggplot")
plt.pie([light, light_medium, medium, medium_heavy, heavy],
        labels=["Under 125lbs", "125-150lbs", "150-175lbs", "175-200lbs", "Over 200lbs"],
        autopct='%.2f %%',
        explode=(.4,.2,0,0,.4) #slice the chart
        )

plt.show()
```
